```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
loans = pd.read_csv('loan_data_final.csv')
loans.head(5)
```

```python
loans.info()
```

```python
loans.describe()
```

```python
plt.figure(figsize=(17,8))
sns.heatmap(loans.corr(),annot=True,cmap='inferno',linewidths=1)
```

```python
plt.figure(figsize=(12,8))
loans[loans['credit.policy']==1]['fico'].hist(bins=45,color='green',label='C.P - 1',alpha=0.6)
loans[loans['credit.policy']==0]['fico'].hist(bins=45,color='blue',label='C.P - 0',alpha=0.6)
plt.xlabel("FICO")
plt.legend()
```
