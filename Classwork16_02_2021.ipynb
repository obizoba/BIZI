```
#defining the model
from sklearn.neighbors import KNeighborsClassifier
#defining data
trian_data=[[2,4],[5,6],[7,6],[8,3]]
train_labels=[[1],[0],[1],[0]]
#test data
##teat inputs
test_data=[[5,6]]
## Validation data
gold_label=[[1]]
#create model
model=KNeighborsClassifier(n_neighbors=2)
model.fit(trian_data,train_labels)
#predict
results=model.predict(test_data)
#Accuracy
print(results)
print(gold_label)
```
